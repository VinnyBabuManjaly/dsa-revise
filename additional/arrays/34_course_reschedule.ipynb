### Problem: Course Schedule

You are given an integer `num_courses` representing the total number of courses, labeled `0` to `num_courses - 1`.

You are also given a list `prerequisites` where each element is `[a, b]`, meaning:
- To take course `a`, you must first take course `b`.

Return **True** if it is possible to finish all courses.
Return **False** otherwise.

#### Example interpretation:
A prerequisite pair `[1, 0]` means:
- To take course 1, you must complete course 0.

#### Constraints:
- 1 ≤ num_courses ≤ 1500  
- 0 ≤ prerequisites.length ≤ 1000  
- prerequisites[i].length = 2  
- 0 ≤ a[i], b[i] < num_courses  
- All prerequisite pairs are unique.
