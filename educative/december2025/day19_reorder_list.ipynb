[Reorder List](https://leetcode.com/problems/reorder-list/description/)

You are given the head of a singly linked-list. The list can be represented as:

L0 → L1 → … → Ln - 1 → Ln
Reorder the list to be on the following form:

L0 → Ln → L1 → Ln - 1 → L2 → Ln - 2 → …
You may not modify the values in the list's nodes. Only nodes themselves may be changed.

Example 1:
Input: head = [1,2,3,4]
Output: [1,4,2,3]

Example 2:
Input: head = [1,2,3,4,5]
Output: [1,5,2,4,3]

Constraints:
The number of nodes in the list is in the range [1, 5 * 104].
1 <= Node.val <= 1000

```python
from typing import Optional

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def reorderList(head: Optional[ListNode]) -> None:
    
    # Finding middle
    slow, fast = head, head

    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next

    # Slow is the middle

    # Reversing second half
    prev, curr = None, slow.next
    slow.next = None

    while curr:
        next= curr.next
        curr.next = prev
        prev = curr
        curr = next

    # curr is the head of reversed list 

    # Merging the lists
    # head and curr are the heads of the two linked lists
    head1, head2 = head, prev
    while head1 and head2:
        next1 = head1.next
        next2 = head2.next

        head1.next = head2
        head2.next = next1

        head1 = next1
        head2 = next2
```

Approach: Split, reverse, then merge two halves

Insight: The list is first split into two halves using slow–fast pointers. The second half is reversed, and then the two halves are merged by alternating nodes from the front and back, which produces the required pattern in-place without changing node values.

Tricky Part: The main care points are breaking the list cleanly at the middle before reversing, and weaving the two halves without losing references, off‑by‑one mistakes here can easily create cycles or drop nodes from the final list.
