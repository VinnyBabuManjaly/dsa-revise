[Nth Magical Number](https://leetcode.com/problems/nth-magical-number/description/)

A positive integer is magical if it is divisible by either a or b.

Given the three integers n, a, and b, return the nth magical number. Since the answer may be very large, return it modulo 10^9 + 7.

Example 1:
Input: n = 1, a = 2, b = 3
Output: 2

Example 2:
Input: n = 4, a = 2, b = 3
Output: 6

Constraints:
1 <= n <= 109
2 <= a, b <= 4 * 104

```python
import math

def nthMagicalNumber(n: int, a: int, b: int) -> int:
    lower = min(a, b)
    upper = n * min(a, b)

    while lower <= upper:
        mid = (lower + upper) // 2
        count = (mid // a) + (mid // b) - mid // math.lcm(a, b)
        
        if count < n:
            lower = mid + 1
        else:
            upper = mid - 1

    return lower % (10**9+7)
```

Approach: Binary search on the answer

Insight: Instead of listing all numbers divisible by a or b, searched directly on the number line. For a given candidate value, counted how many numbers up to that value are divisible by a or b, adjusting for overlap using their common multiples. Then narrowed the range until landing exactly on the position where the count first reaches n.

Tricky Part: The main care point is correctly handling the overlap between multiples of a and b so they are not double-counted, and keeping the search bounds tight and safe for large n. The final result is taken modulo 1e9+7, as required by the problem.
