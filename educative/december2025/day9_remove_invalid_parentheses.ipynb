[Remove Invalid Parentheses](https://leetcode.com/problems/remove-invalid-parentheses/)

Given a string s that contains parentheses and letters, remove the minimum number of invalid parentheses to make the input string valid.

Return a list of unique strings that are valid with the minimum number of removals. You may return the answer in any order.

Example 1:   
Input: s = "()())()"  
Output: ["(())()","()()()"]
 
Example 2:   
Input: s = "(a)())()"   
Output: ["(a())()","(a)()()"]  

Example 3:   
Input: s = ")("   
Output: [""]

Constraints:   
1 <= s.length <= 25   
s consists of lowercase English letters and parentheses '(' and ')'.   
There will be at most 20 parentheses in s.   

```python
from collections import deque

def removeInvalidParentheses(s: str) -> list[str]:

    def isValid(string):
        balance = 0
        for ch in string:
            if ch == '(': balance += 1
            if ch == ')': balance -= 1
            if balance < 0: return False
        return balance == 0

    queue = deque([s])
    visited = set()
    found = False
    result = []

    # Checking string validity
    while queue:

        for _ in range(len(queue)):
            string = queue.popleft()

            if isValid(string):
                result.append(string)
                found = True
            else:
                if found:
                    continue

                for i in range(len(string)):
                    if string[i] in '()':
                        next_str = string[:i]+string[i+1:]
                        if next_str not in visited:
                            visited.add(next_str)
                            queue.append(next_str)

        if found:
            break

    return result
```

Approach: BFS over all strings with minimal removals

Insight: Start from the original string and, level by level, remove one parenthesis at each position to generate new candidates. As soon as a level produces valid strings (balanced parentheses), collect all of them and stop going deeperâ€”this guarantees the minimum number of removals.

Tricky Part: Two small but important points: avoid revisiting the same string by tracking a visited set, and stop generating next-level strings once at least one valid string is found, otherwise extra work is done and more than the minimum removals may be included.
