[Longest Happy String](https://leetcode.com/problems/longest-happy-string/description/)

A string s is called happy if:

1. s only contains 'a', 'b', and 'c'.
2. s does not contain "aaa", "bbb", or "ccc".
3. s contains at most a occurrences of 'a'.
4. s contains at most b occurrences of 'b'.
5. s contains at most c occurrences of 'c'.

Given integers a, b, and c, return the longest possible happy string.  
If multiple longest strings exist, return any of them.  
If no valid string exists, return "".

Examples:

Input: a = 1, b = 1, c = 7  
Output: "ccaccbcc"

Input: a = 7, b = 1, c = 0  
Output: "aabaa"

Constraints:
0 <= a, b, c <= 100  
a + b + c > 0


```python
import heapq

def longestDiverseString(a: int, b: int, c: int) -> str:
    happy = ""
    maxHeap = []
    if a > 0: heapq.heappush(maxHeap, (-a, 'a'))
    if b > 0: heapq.heappush(maxHeap, (-b, 'b'))
    if c > 0: heapq.heappush(maxHeap, (-c, 'c'))
    
    while maxHeap:
        count1, ch1 = heapq.heappop(maxHeap)

        # Correct triple check
        if len(happy) >= 2 and happy[-1] == ch1 and happy[-2] == ch1:
            # Need to take second most frequent
            if not maxHeap:
                break  # no alternative → stop

            count2, ch2 = heapq.heappop(maxHeap)
            happy += ch2
            count2 += 1

            if count2 < 0:
                heapq.heappush(maxHeap, (count2, ch2))

            # Push ch1 back unchanged
            heapq.heappush(maxHeap, (count1, ch1))

        else:
            # Safe to use ch1
            happy += ch1
            count1 += 1

            if count1 < 0:
                heapq.heappush(maxHeap, (count1, ch1))

    return happy
```

Approach: Greedy choice with a max-heap

Insight: Always pick the character with the highest remaining count, unless it would create three in a row, in which case temporarily switch to the next-best character. After using a character, push it back with an updated remaining count so the choice stays fair and balanced across a, b, and c.

Tricky Part: The main care point is handling the “three in a row” situation correctly—when the top character would violate the rule, it must be skipped for that step, and if no other character is available at that moment, the string is complete and the process stops.
