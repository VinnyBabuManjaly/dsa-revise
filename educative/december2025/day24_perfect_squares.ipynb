```python
def numSquares(n: int) -> int:
    squares = []
    i = 1
    while i * i <= n:
        squares.append(i * i)
        i += 1

    dp = [float('inf')] * (n + 1)
    dp[0] = 0

    for x in range(1, n + 1):
        for s in squares:
            if s > x:
                break
            dp[x] = min(dp[x], 1 + dp[x - s])

    return dp[n]
```

Core Idea: Build the answer bottom-up with dynamic programming

The Pattern: Unbounded coin-change over all perfect squares
Precompute all perfect squares up to n, then for every value from 1 to n, choose the best way to form it by adding one square on top of a previously solved smaller value.

Why It Works Here:
Each number up to n can be seen as “current square + a smaller remainder already solved.” Reusing these smaller results avoids recomputing combinations and guarantees the minimal count of squares for every value up to n.

Production Lesson:
This pattern matches many real workloads where a target cost, latency, or capacity is assembled from reusable building blocks—solve smaller subproblems once, cache them, and scale the result instead of recomputing from scratch.

Edge Worth Remembering:
Initializing the base case correctly (0 requires 0 squares) and skipping squares larger than the current target keeps the transitions simple and the solution efficient.
