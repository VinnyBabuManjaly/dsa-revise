Max Consecutive Ones

Given a binary array nums, return the maximum number of consecutive 1's in the array.

Example 1:

Input:
nums = [1,1,0,1,1,1]

Output:
3

Explanation:
The first two digits or the last three digits are consecutive 1s. The maximum number of consecutive 1s is 3.

Example 2:

Input:
nums = [1,0,1,1,0,1]

Output:
2

Constraints:
1 <= nums.length <= 10^5
nums[i] is either 0 or 1

```python
def findMaxConsecutiveOnes(nums: list[int]) -> int:
    current, maxones = 0, 0

    for num in nums:
        if num == 1:
            current += 1
            if current > maxones:
                maxones = current
        else:
            current = 0
    
    return maxones
```

Approach: Single pass with running streak count

Insight: Walk through the array once, keep a running count of how many 1s are seen in a row, and reset this count whenever a 0 appears. Track the best streak seen so far; that value is the answer.

Tricky Part: The only small detail is to update the maximum streak at the right time, including when the last element is 1, so the final run of 1s is not missed.
