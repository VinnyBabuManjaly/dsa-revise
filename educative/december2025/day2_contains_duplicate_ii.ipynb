[Contains Duplicate II](https://leetcode.com/problems/contains-duplicate-ii/description/)

Given an integer array nums and an integer k, return true if there are two distinct indices i and j in the array such that nums[i] == nums[j] and abs(i - j) <= k.

Example 1:
Input: nums = [1,2,3,1], k = 3
Output: true

Example 2:
Input: nums = [1,0,1,1], k = 1
Output: true

Example 3:
Input: nums = [1,2,3,1,2,3], k = 2
Output: false

Constraints:
1 <= nums.length <= 10^5
-10^9 <= nums[i] <= 10^9
0 <= k <= 10^5

```python
def contains_nearby_duplicate(nums: list[int], k: int) -> bool:
    
    nummap = {}

    for i, num in enumerate(nums):
        if num in nummap and i - nummap[num] <= k:
            return True
        nummap[num] = i
    
    return False
```

Approach: Hash map of last seen index

Insight: Walk through the array once, keeping the last index for each value. When the same value appears again, just check if the gap between the two indices is at most k; if not, update the stored index and continue.

Tricky Part: The important detail is updating the index in the map every time a value is seen so the distance check always uses the closest previous occurrence, not an older one.
