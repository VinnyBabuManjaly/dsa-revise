# Find Longest Self-Contained Substring

You are given a string `s`, consisting of lowercase English letters.

Your task is to find the length of the longest self-contained substring of `s`.

## Definition: Self-Contained Substring

A substring `t` of `s` is self-contained if:

1. `t` is not equal to the entire string `s`
2. Every character in `t` appears only inside `t`  
   (i.e., none of its characters appear elsewhere in `s`)

This means all characters in the substring are exclusive to that substring and appear exactly once in the entire string.

Return the maximum length of such a substring.  
If no valid substring exists, return `-1`.

## Constraints
- `2 ≤ s.length ≤ 1000`
- `s` consists only of lowercase letters

## Examples

### Example 1
Input:  
`s = "abcaefg"`

- `a` appears twice → cannot be in the substring  
- `e, f, g` appear only once

Valid self-contained substring: `"efg"`  
Output: `3`

### Example 2
Input:  
`s = "abcabd"`

- `a, b, c` appear twice  
- `d` appears once

Valid substring: `"d"`  
Output: `1`

### Example 3
Input:  
`s = "aabbcc"`

All characters appear more than once.

Output: `-1`

### Example 4
Input:  
`s = "abcd"`

All characters are unique, but substring cannot be the entire string.

Valid substrings: `"abc"`, `"bcd"`  
Output: `3`

```python
from collections import defaultdict

def max_substring_length(s):
    max_len = -1
    n = len(s)

    positions = defaultdict(list)
    for i, ch in enumerate(s):
        if ch not in positions:
            positions[ch] = [i, i]
        else:
            positions[ch][1] = i 

    window_start = 0
    for ch in positions:
        window_start = positions[ch][0]
        window_end = positions[ch][1]
        j = window_start

        while j < n:
            if positions[s[j]][0] < window_start:
                break
            window_end = max(window_end, positions[s[j]][1])
        
            length = window_end-window_start+1
            if window_end == j and length != n:
                max_len = max(max_len, length)
            j += 1


    return max_len

max_substring_length(s = "zzzzzz")
```
