Happy Number

Write an algorithm to determine if a number n is happy.

A happy number is defined by this process:
- Start with any positive integer.
- Replace the number by the sum of the squares of its digits.
- Repeat until the number becomes 1 (happy) or enters a loop that does not include 1 (not happy).

Return true if n is a happy number, otherwise false.

Example 1:
Input: n = 19
Output: true

Explanation:
1² + 9² = 82
8² + 2² = 68
6² + 8² = 100
1² + 0² + 0² = 1

Example 2:
Input: n = 2
Output: false

Constraints:
1 <= n <= 2^31 - 1


```python
def is_happy_number(n):
    slow, fast = n, n

    while True:
        slow = next_num(slow)
        fast = next_num(next_num(fast))

        if slow == 1 or fast == 1:
            return True

        if slow == fast:
            return False

def next_num(n):
    s = 0
    while n > 0:
        digit = n % 10
        s += digit * digit
        n //= 10
    return s
```

Approach: Floyd's Cycle Detection 

Insight: Transform the number repeatedly by summing squares of its digits. Use slow and fast pointers moving at different speeds—if they meet, there's a cycle (unhappy). If either reaches 1, it's happy. Simple arithmetic reveals whether a number converges or loops.

Reflection: Cycle detection elegantly solves many real-world problems—from deadlock detection in systems to identifying fraudulent patterns in data.
