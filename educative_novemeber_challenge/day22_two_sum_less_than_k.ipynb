[Two Sum Less Than K](https://leetcode.com/problems/two-sum-less-than-k/description/)

Given an array nums and an integer k, return the maximum sum of any two distinct elements such that the sum is less than k. If no such pair exists, return -1.

Examples:

Example 1:
Input: nums = [34,23,1,24,75,33,54,8], k = 60
Output: 58
Explanation: The pair (34,24) has the maximum sum less than 60.

Example 2:
Input: nums = [10,20,30], k = 15
Output: -1
Explanation: No two numbers sum to less than 15.

Example 3:
Input: nums = [5,1,3,4,7], k = 12
Output: 11
Explanation: The pair (4,7) gives sum 11 which is the largest below 12.

Example 4:
Input: nums = [2,9,1,8], k = 10
Output: 9
Explanation: The pair (1,8) gives sum 9.

Constraints:
1 <= nums.length <= 1000
1 <= nums[i] <= 1000
1 <= k <= 2000

```python
def two_sum_less_than_k(nums: list, k:int) -> int:

    nums.sort()
    left, right = 0, len(nums)-1
    minsum = -1

    while left < right:
        twosum = nums[left] + nums[right]
        if twosum >= k:
            right -= 1
        else: 
            minsum = max(minsum, twosum)
            left += 1

    return minsum
```
