## [Minimum Cost to Connect Sticks](https://leetcode.com/problems/minimum-cost-to-connect-sticks/description/)  

You are given an array sticks where sticks[i] is the length of the ith stick. You can connect any two sticks of lengths x and y into one stick by paying a cost of x + y. You must repeat this process until there is only one stick left. The goal is to find the minimum total cost to connect all the sticks.  

Return the minimum cost to connect all the sticks.  

Constraints:  
1 <= sticks.length <= 10000  
1 <= sticks[i] <= 10000

Example 1:  
Input: sticks = [2,4,3]  
Output: 14  
Explanation:  
Connect 2 and 3 into 5 → cost = 5  
Now connect 5 and 4 into 9 → cost = 9  
Total cost = 5 + 9 = 14

Example 2:  
Input: sticks = [1,8,3,5]  
Output: 30

Example 3:  
Input: sticks = [5]  
Output: 0  
Explanation: No cost if only one stick.

```python
import heapq

def connect_sticks(sticks: list[int]) -> int:
    
    heapq.heapify(sticks)
    mincost = 0

    while sticks and len(sticks) > 1:
        cost = heapq.heappop(sticks) + heapq.heappop(sticks)
        mincost += cost
        heapq.heappush(sticks, cost)

    return mincost
```

Approach: Greedy min-heap combination

Insight: Always combine the two smallest sticks using a min-heap, ensuring each join is as cheap as possible and the running total is minimized.
