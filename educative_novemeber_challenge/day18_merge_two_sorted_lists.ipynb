SortingMerge Two Sorted Lists

Easy

You are given the heads of two sorted linked lists list1 and list2.

Merge the two lists into one sorted list. The list should be made by splicing together the nodes of the first two lists. 

Return the head of the merged linked list.

Example 1:  
Input: list1 = [1,2,4], list2 = [1,3,4]  
Output: [1,1,2,3,4,4]  
Example 2:  
Input: list1 = [], list2 = []  
Output: []  
Example 3:  
Input: list1 = [], list2 = [0]  
Output: [0]

Constraints:  
The number of nodes in both lists is in the range [0, 50].  
-100 <= Node.val <= 100  
Both list1 and list2 are sorted in non-decreasing order.

```python

from linked_list_node import LinkedListNode

class LinkedListNode:
    def __init__(self, data, next=None):
        self.data = data
        self.next = next

def merge_two_lists(head1:LinkedListNode, head2: LinkedListNode):  
    
    dummy = LinkedListNode(-1)
    current = dummy

    while head1 and head2:
        if head1.data <= head2.data:
            current.next = head1
            head1 = head1.next

        else:
            current.next = head2
            head2 = head2.next        

        current = current.next

    current.next = head1 if head1 else head2

    return dummy.next
        
```

Approach: Two-pointer traversal with a dummy node.

Insight: By advancing pointers based on value and directly linking nodes, the solution efficiently splices together the originals with O(n+m) time and O(1) spaceâ€”no extra allocations.
