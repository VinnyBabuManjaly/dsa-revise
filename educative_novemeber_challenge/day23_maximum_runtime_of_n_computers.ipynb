## [Maximum Running Time of N Computers](https://leetcode.com/problems/maximum-running-time-of-n-computers/description/)

You have n computers and an array batteries where batteries[i] is the number of minutes the ith battery can run a computer. Each computer can have at most one battery inserted at any time, but you are allowed to swap batteries instantly at any integer time step. Batteries cannot be recharged.

Your task is to determine the maximum number of minutes you can run all n computers simultaneously.

Example 1:
Input: n = 2, batteries = [3,3,3]
Output: 4

Example 2:
Input: n = 2, batteries = [1,1,1,1]
Output: 2

Constraints:
1 <= n <= batteries.length <= 100000
1 <= batteries[i] <= 1e9


```python
def maxRunTime(batteries: list[int], n: int) -> int:
    
    low, high = 0, sum(batteries)//n

    while low <= high:
        mid = (low+high)//2
        
        # Maximum possible minutes possible to run with all batteries
        maxminutes = sum(min(battery, mid) for battery in batteries)
        if maxminutes >= mid * n:
            # Higher values are possible
            low = mid + 1
        else: 
            high = mid - 1

    return high
```

Approach: Binary Search on running time

Insight: By leveraging binary search over possible running times and shuffling batteries to equalize usage, the solution pinpoints the maximum sustainable time without trial-and-error simulation.
