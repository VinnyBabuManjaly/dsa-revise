# Letter Combinations of a Phone Number

You are given a string digits made up of digits from 2 through 9 inclusive.

Each digit (not including 1) is mapped to a set of characters as shown below:

A digit could represent any one of the characters it maps to.

Return all possible letter combinations that digits could represent. You may return the answer in any order.

Example 1:  
Input: digits = "34"   
Output: ["dg","dh","di","eg","eh","ei","fg","fh","fi"]  

Example 2:   
Input: digits = ""   
Output: []  

Constraints:   
0 <= digits.length <= 4   
2 <= digits[i] <= 9

```python
class Solution:
    def letterCombinations(self, digits: str) -> list[str]:

        lettermap = {2: "abc", 3:"def", 4:"ghi", 5:"jkl", 6:"mno", 7:"qprs", 8:"tuv", 9:"wxyz"}
        result = []
        n = len(digits)

        substring = []
        def dfs(i):
            if i >= n:
                result.append("".join(substring))
                return

            for s in lettermap[int(digits[i])]:
                substring.append(s)
                dfs(i+1)
                substring.pop()

        if digits:
            dfs(0)

        return result
```

### Approach: Backtracking (Digit-to-Letter Mapping)

**Main Logic:**

* Create a mapping from digit to corresponding letters (phone keypad).
* Use DFS to build combinations character by character.
* For each digit, iterate through its mapped letters.
* Add one letter, move to next digit, and continue.
* When all digits are processed, add the built string to result.
* Backtrack to explore other letter choices.
* If input is empty, return empty list.

**Key idea:**
For each digit, choose one possible letter and recursively build all combinations.



**Time Complexity**: O(n.4ⁿ)
Each digit can map to at most 4 letters.

**Space Complexity**: O(n)
Recursion depth equals number of digits.

* This is a classic Cartesian product problem.
* Backtracking handles multi-choice branching naturally.
* Each level of recursion represents one decision position.
