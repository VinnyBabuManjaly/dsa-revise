# Word Search

Given a 2-D grid of characters board and a string word, return true if the word is present in the grid, otherwise return false.

For the word to be present it must be possible to form it with a path in the board with horizontally or vertically neighboring cells. The same cell may not be used more than once in a word.

Example 1:  
Input:    
board = [    
  ["A","B","C","D"],   
  ["S","A","A","T"],   
  ["A","C","A","E"]   
],   
word = "CAT"   
Output: true

Example 2:    
Input:    
board = [   
  ["A","B","C","D"],   
  ["S","A","A","T"],   
  ["A","C","A","E"]   
],   
word = "BAT"   
Output: false

Constraints:   
1 <= board.length, board[i].length <= 5   
1 <= word.length <= 10   
board and word consists of only lowercase and uppercase English letters.

```python
class Solution:
    def solveNQueens(self, n: int) -> list[list[str]]:
        cols, posDiag, negDiag = set(), set(), set()

        result = []
        board = [["."] * n for _ in range(n)]
        print(board)

        def backtrack(r):
            if r == n:
                copy = ["".join(row) for row in board]
                result.append(copy)
                return

            for c in range(n):
                if c in cols or r+c in posDiag or r-c in negDiag:
                    continue
                
                cols.add(c)
                posDiag.add(r+c)
                negDiag.add(r-c)
                board[r][c] = "Q"

                backtrack(r+1)

                cols.remove(c)
                posDiag.remove(r+c)
                negDiag.remove(r-c)
                board[r][c] = "."

        backtrack(0)
        return result
```

## Approach: Backtracking with Column and Diagonal Tracking

**Main Logic:**

* Place queens row by row.
* For each row, try placing a queen in every column.
* Before placing, check if:

  * Column is already used.
  * Positive diagonal `(r + c)` is used.
  * Negative diagonal `(r - c)` is used.
* If safe, place the queen and mark column and diagonals.
* Recursively move to the next row.
* If all rows are filled, store the board configuration.
* Backtrack by removing the queen and clearing the markers.

**Key idea:**
Track columns and diagonals using sets so invalid placements are skipped immediately.





**Time Complexity**: O(N!)
Each row tries multiple column placements.

**Space Complexity**: O(N)
Recursion stack and tracking sets.

### What can be learned from this problem?

* Constraint tracking drastically reduces search space.
* Diagonal relationships can be represented using `r+c` and `r-c`.
* Backtracking is ideal for board placement and constraint problems.
