# Palindrome Partitioning

Given a string s, split s into substrings where every substring is a palindrome. Return all possible lists of palindromic substrings.

You may return the solution in any order.

Example 1:   
Input: s = "aab"   
Output: [["a","a","b"],["aa","b"]]   

Example 2:  
Input: s = "a"   
Output: [["a"]]   

Constraints:   
1 <= s.length <= 20   
s contains only lowercase English letters.

```python
class Solution:
    def partition(self, s: str) -> list[list[str]]:

        result = []
        subset = []
        n = len(s)

        def dfs(i):
            if i >= n:
                result.append(subset.copy())
                return

            for j in range(i, n):
                if self.isPalindrome(s[i:j+1]):
                    subset.append(s[i:j+1])
                    dfs(j+1)
                    subset.pop()

        dfs(0)
        return result


    def isPalindrome(self, string):
        left, right = 0, len(string)-1

        while left < right:
            if string[left] != string[right]:
                return False
            left, right = left+1, right-1

        return True
```

### Approach: Backtracking with Palindrome Checking

**Main Logic:**

* Start from index `0` and try all possible substring partitions.
* For each position, expand the substring forward.
* Check if the current substring is a palindrome.
* If it is a palindrome, include it in the current partition.
* Recursively continue partitioning the remaining string.
* Backtrack by removing the last added substring and try other splits.
* When the end of the string is reached, store the valid partition.

**Key idea:**
Build partitions step-by-step and only continue recursion when the chosen substring is a palindrome.


**Time Complexity**: O(N × 2^N)
Each position can create multiple partitions and palindrome checking takes linear time.

**Space Complexity**: O(N)
Recursion depth and temporary subset storage.
