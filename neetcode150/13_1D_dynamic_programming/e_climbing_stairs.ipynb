Climbing Stairs

You are given an integer n representing the number of steps to reach the top of a staircase. You can climb with either 1 or 2 steps at a time.

Return the number of distinct ways to climb to the top of the staircase.

Example 1:  
Input: n = 2  
Output: 2   
Explanation:   
1 + 1 = 2   
2 = 2   

Example 2:  
Input: n = 3   
Output: 3   
Explanation:   
1 + 1 + 1 = 3   
1 + 2 = 3   
2 + 1 = 3

Constraints:   
1 <= n <= 30

```python
class Solution:
    def climbStairs(self, n: int) -> int:
        one, two = 1, 1
        
        for i in range(n-2, -1, -1):
            temp = one + two
            one, two = temp, one

        return one
```

**Approach**: Dynamic Programming (Bottom-Up with Constant Space)

Main Logic:
- Start from the top of the stairs where there is exactly one way to stand.
- Keep track of the number of ways to reach the next step and the step after that.
- Move backward step by step.
- At each step, total ways are obtained by adding ways from the next two steps.
- Update the variables as you move.
- Return the total ways to reach the first step.

Key idea:
The number of ways to reach a step depends only on the previous two steps, so extra memory is not needed.

**Time Complexity**: O(n)
Each step is processed once.

**Space Complexity**: O(1)
Only two variables are used to store previous results.
