House Robber

You are given an integer array nums where nums[i] represents the amount of money the ith house has. The houses are arranged in a straight line, i.e. the ith house is the neighbor of the (i-1)th and (i+1)th house.

You are planning to rob money from the houses, but you cannot rob two adjacent houses because the security system will automatically alert the police if two adjacent houses were both broken into.

Return the maximum amount of money you can rob without alerting the police.

Example 1:  
Input: nums = [1,1,3,3]  
Output: 4   
Explanation: nums[0] + nums[2] = 1 + 3 = 4.   

Example 2:   
Input: nums = [2,9,8,3,6]    
Output: 16    
Explanation: nums[0] + nums[2] + nums[4] = 2 + 8 + 6 = 16.    

Constraints:   
1 <= nums.length <= 100   
0 <= nums[i] <= 100   

```python
class Solution:
    def rob(self, nums: list[int]) -> int:

        rob1, rob2 = 0, 0

        for n in nums:
            temp = max(n+rob1, rob2)
            rob1, rob2 = rob2, temp

        return rob2
```

**Approach**: Dynamic Programming (Greedy with Rolling Variables)

Main Logic:
Start with two variables to track the best loot so far.
At each house, decide whether to rob it or skip it.
If you rob the current house, add its money to the best result from two houses ago.
If you skip it, keep the best result from the previous house.
Choose the maximum of these two options.
Move forward updating the variables.
Return the final maximum amount.

Key idea:
At every step, you choose the best outcome without storing all previous results, since only the last two states matter.

**Time Complexity**: O(n)
Each house is visited once.

**Space Complexity**: O(1)
Only two variables are used regardless of input size.
