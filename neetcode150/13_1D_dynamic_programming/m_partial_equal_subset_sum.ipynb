Partition Equal Subset Sum

You are given an array of positive integers nums.

Return true if you can partition the array into two subsets, subset1 and subset2 where sum(subset1) == sum(subset2). Otherwise, return false.

Example 1:  
Input: nums = [1,2,3,4]  
Output: true   
Explanation: The array can be partitioned as [1, 4] and [2, 3].

Example 2:   
Input: nums = [1,2,3,4,5]    
Output: false    

Constraints:   
1 <= nums.length <= 100   
1 <= nums[i] <= 50

```python
class Solution:
    def canPartition(self, nums: list[int]) -> bool:
        if sum(nums) % 2:
            return False

        target = sum(nums) // 2
        n = len(nums)
        dp = set()
        dp.add(0)

        for i in range(n-1, -1,-1):
            newset = set()
            for j in dp:
                sumvalue = nums[i] + j
                if sumvalue == target:
                    return True
                newset.add(sumvalue)
                newset.add(j)
            dp = newset

        return False
```

**Approach**: Dynamic Programming using Set (Subset Sum)

Main Logic:
- First check if the total sum is odd.
    - If it is odd, it cannot be split into two equal subsets.
- Set the target as half of the total sum.
- Use a set to store all possible subset sums seen so far.
- Start with 0 as the initial possible sum.
- Iterate through the numbers one by one.
- For each number, create new possible sums by including or excluding it.
    - If any sum equals the target, return True.
    - If all numbers are processed without reaching the target, return False.

Key idea:
If a subset with sum equal to half of the total exists, the array can be partitioned into two equal subsets.

**Time Complexity**: O(n Ã— target)   
In the worst case, all sums up to the target are explored.

**Space Complexity**: O(target)
The set stores possible subset sums up to the target.
