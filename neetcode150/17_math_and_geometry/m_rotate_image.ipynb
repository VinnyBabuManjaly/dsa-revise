# Rotate Image

Given a square n x n matrix of integers matrix, rotate it by 90 degrees clockwise.

You must rotate the matrix in-place. Do not allocate another 2D matrix and do the rotation.

Example 1:   
Input: matrix = [   
  [1,2],   
  [3,4]   
]   
Output: [    
  [3,1],    
  [4,2]   
]   

Example 2:   
Input: matrix = [   
  [1,2,3],   
  [4,5,6],   
  [7,8,9]   
]   
Output: [    
  [7,4,1],   
  [8,5,2],   
  [9,6,3]    
]

Constraints:   
n == matrix.length == matrix[i].length    
1 <= n <= 20    
-1000 <= matrix[i][j] <= 1000   

```python
class Solution:
    def rotate(self, matrix: list[list[int]]) -> None:
        
        left, right = 0, len(matrix) - 1

        while left < right:

            top, bottom = left, right

            for i in range(right-left):

                # keeping top left 
                topLeft = matrix[top][left + i]

                # bottom left to top left
                matrix[top][left + i] = matrix[bottom - i][left]

                # bottom right to bottom left
                matrix[bottom - i][left] = matrix[bottom][right - i]

                # top right to bottom right
                matrix[bottom][right - i] = matrix[top + i][right]

                # top left to top right
                matrix[top + i][right] = topLeft

            left += 1
            right -=1
```

### Approach: Layer-by-Layer In-place Rotation (4-way Swap)

**Main Logic:**

* Rotate the matrix layer by layer from outer to inner.
* Use two pointers `left` and `right` to track current square layer.
* For each layer:

  * Iterate across the top row.
  * Perform a 4-way swap:

    * bottom-left → top-left
    * bottom-right → bottom-left
    * top-right → bottom-right
    * stored top-left → top-right
* Move inward by updating `left += 1` and `right -= 1`.
* Continue until all layers are processed.

**Key idea:**
Rotate elements in groups of four to perform in-place 90° rotation without extra space.





**Time Complexity**: O(n²)
Every element in the matrix is visited once.

**Space Complexity**: O(1)
Rotation is done in-place with constant extra memory.
