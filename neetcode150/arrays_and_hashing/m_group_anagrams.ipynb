```python
from collections import defaultdict

def group_anagrams(strs: list):
    anagrams = defaultdict(list)

    for string in strs:
        charmap = [0] * 26
        for char in string:
            charmap[ord(char)-ord('a')] += 1

        anagrams[tuple(charmap)].append(string)
    
    return anagrams.values()

group_anagrams(["duel", "dule", "speed"])
```
