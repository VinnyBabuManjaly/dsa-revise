# Contains Duplicate

Given an integer array nums, return true if any value appears more than once in the array, otherwise return false.

Example 1:   
Input: nums = [1, 2, 3, 3]   
Output: true

Example 2:   
Input: nums = [1, 2, 3, 4]   
Output: false

```python
from typing import List

class Solution:
    def hasDuplicate(self, nums: List[int]) -> bool:
        return len(nums) > len(set(nums))
```

### Approach: Set Length Comparison

**Main Logic:**

* Convert the list to a set, which automatically removes duplicates
* Compare the length of the original list with the length of the set
* If the list has more elements than the set, there must be duplicates

**Key idea:**
Leverage Python's built-in set data structure which only stores unique elements.

**Time Complexity**: O(n)
Converting a list to a set requires iterating through all n elements.

**Space Complexity**: O(n)
Creating a set requires storing up to n elements in the worst case (when all elements are unique).
