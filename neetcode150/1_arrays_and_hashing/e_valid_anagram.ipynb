# Valid Anagram

Given two strings s and t, return true if the two strings are anagrams of each other, otherwise return false.

An anagram is a string that contains the exact same characters as another string, but the order of the characters can be different.

Example 1:   
Input: s = "racecar", t = "carrace"   
Output: true   

Example 2:   
Input: s = "jar", t = "jam"   
Output: false

Constraints:   
s and t consist of lowercase English letters.

```python
class Solution:
    def isAnagram(self, s: str, t: str) -> bool:
        if len(s) != len(t):
            return False

        smap, tmap = {}, {}

        for i in range(len(s)):
            smap[s[i]] = smap.get(s[i], 0) + 1
            tmap[t[i]] = tmap.get(t[i], 0) + 1

        return smap == tmap
```

### Approach: Frequency Count using Hash Maps

**Main Logic:**

* If lengths of both strings are different → they cannot be anagrams.
* Use two hash maps (dictionaries):

  * One to count frequency of characters in string `s`.
  * Another to count frequency of characters in string `t`.
* Traverse both strings:

  * Increment count for each character in respective maps.
* Compare both maps:

  * If they are equal → strings are anagrams.
  * Else → not anagrams.

**Key idea:**
Two strings are anagrams if they have identical character frequencies.




**Time Complexity**: O(n)
Single pass to build maps and compare.

**Space Complexity**: O(1)
At most 26 lowercase English letters (constant space).

### What can be learned from this problem?

* Hash maps are powerful for counting problems.
* Equal frequency → strong indicator of equivalence.
* Avoid sorting when linear-time solutions exist.
* Pattern applies to many string comparison problems (permutations, substrings, etc.).
