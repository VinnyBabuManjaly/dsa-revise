Merge Intervals

Medium
Given an array of intervals where intervals[i] = [starti, endi], merge all overlapping intervals, and return an array of the non-overlapping intervals that cover all the intervals in the input.

Example 1:

Input: intervals = [[1,3],[2,6],[8,10],[15,18]]
Output: [[1,6],[8,10],[15,18]]
Explanation: Since intervals [1,3] and [2,6] overlap, merge them into [1,6].
Example 2:

Input: intervals = [[1,4],[4,5]]
Output: [[1,5]]
Explanation: Intervals [1,4] and [4,5] are considered overlapping.

Constraints:
1 <= intervals.length <= 104
intervals[i].length == 2
0 <= starti <= endi <= 104

```python
class Solution:
    def mergeIntervals(self, intervals: list[list[int]]) -> list[list[int]]:
        intervals.sort()
        result = [intervals[0]]

        for startnew, endnew in intervals[1:]:
            end = result[-1][1]
            if end >= startnew:
                result[-1][1] = max(end, endnew)
            else:
                result.append([startnew, endnew])

        return result
```

**Approach**: Sorting + Greedy Interval Merging

Main Logic:
- Sort all intervals by their starting time.
- Initialize the result list with the first interval.
- Iterate through the remaining intervals one by one.
- Compare the current interval with the last interval in the result.
    - If they overlap, merge them by extending the end time.
    - If they do not overlap, add the current interval as a new entry.
- Return the merged result list.

Key idea:
Once intervals are sorted, overlapping intervals will always appear next to each other, allowing simple linear merging.

**Time Complexity**: O(n log n)
Sorting the intervals dominates the runtime.

**Space Complexity**: O(n)
Result list may store all intervals in the worst case.
