[Max Area of Island](https://leetcode.com/problems/max-area-of-island/description/)

You are given a matrix grid where grid[i] is either a 0 (representing water) or 1 (representing land).

An island is defined as a group of 1's connected horizontally or vertically. You may assume all four edges of the grid are surrounded by water.

The area of an island is defined as the number of cells within the island.

Return the maximum area of an island in grid. If no island exists, return 0.   

Example 1:   
Input: grid = [
  [0,1,1,0,1],
  [1,0,1,0,1],
  [0,1,1,0,1],
  [0,1,0,0,1]
]   
Output: 6   
Explanation: 1's cannot be connected diagonally, so the maximum area of the island is 6.   

Constraints:   
1 <= grid.length, grid[i].length <= 50

```python
class Solution:
    def maxAreaOfIsland(self, grid: list[list[int]]) -> int:
        m, n = len(grid), len(grid[0])
        visit = set()

        def _dfs(row, col):
            if row >= 0 and row < m and col >= 0 and col < n and grid[row][col] == 1 and (row, col) not in visit:
                visit.add((row, col))
                return (1 + _dfs(row-1, col) + _dfs(row+1, col) + _dfs(row, col-1) + _dfs(row, col+1))
            else:
                return 0

        area = 0
        for row in range(m):
            for col in range(n):
                area = max(area, _dfs(row, col))

        return area
```

**Approach**: DFS Flood Fill (Connected Component Area)

Main Logic:
- Scan every cell in the grid.
- When a land cell `1` that is not visited is found, start DFS from it.
- Count the current cell as part of the island.
- Recursively explore all four directions to include all connected land.
- Keep track of visited cells to avoid double counting.
- Update the maximum area after each DFS completes.

Key idea:
Each DFS explores one entire island and returns its total size, and the largest such size is the answer.

**Time Complexity**: O(m * n)
Each cell is visited at most once.

**Space Complexity**: O(m * n)
Worst case recursion stack and visited set when the entire grid is land.
